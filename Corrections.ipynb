## машинное обучение подходящей кривой
Необходимо ограничить область допустимых температур или подобрать новый полином

```python
x = np.array((60, 120,200,300,540, 800, 1200, 1600, 1800, 1900))
y = np.array((8, 19,26,36,44, 47, 49, 50, 51, 51))

d = 5
theta = np.polyfit(x, y, deg= d)
model = np.poly1d(theta)
x2 = [i*5 for i in range(450)]
plt.plot(x, y, 'ro')
plt.plot(x2, model(x2))
model
```

## Результат температурной поправки

```python
def upscaleTemperature(t):
    t = t + 2.20364585e-14*t**5 +  -1.35450975e-10*t**4 + 3.22201667e-07*t**3 + -3.72054524e-04*t**2 + 2.14210916e-01*t -3.09947717e+00
    return t
tmperatures = []
for i in range(1000):
    tmperatures.append(upscaleTemperature(i))

plt.plot(tmperatures)
plt.plot([i for i in range(1000)])
plt.show()
```
